```python
import sectio
import os
from sectio.exporters.svg_exporter import save_geometry_to_svg

# Your specific list of profiles
target_profiles = [
    'L70x70x6', 'L35x35x4', 'L50x50x5', 'L90x90x8', 'L45x45x5',
    'L45x45x4', 'L120x120x8', 'L140x140x10', 'L50x50x4', 'L120x120x10',
    'L55x55x5', 'L60x60x5', 'L110x110x8', 'L40x40x4', 'L70x70x8',
    'L75x75x5', 'L60x60x6', 'L55x55x6'
]

output_folder = "exported_svgs"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

print(f"Ready to export {len(target_profiles)} profiles to '{output_folder}/'")
```

```python
# Assuming the family is "L" based on the profile names
family_name = "L"
table_name = f"sections_le"

success_count = 0
failed_profiles = []

for section_id in target_profiles:
    try:
        # 1. Fetch Geometry
        geom = sectio.db.get_section(table_name, section_id)

        # 2. Sanitize Filename
        safe_id = section_id.replace("/", "_").replace(" ", "_")
        filename = f"{safe_id}.svg"

        # 3. Export
        save_geometry_to_svg(
            geom.polygon,
            filename=filename,
            folder=output_folder,
            color="#2ecc71"
        )

        print(f"✅ Exported: {section_id}")
        success_count += 1

    except Exception as e:
        print(f"❌ Failed: {section_id} | Error: {e}")
        failed_profiles.append(section_id)

print(f"\n--- Export Complete ---")
print(f"Successfully exported: {success_count}/{len(target_profiles)}")

if failed_profiles:
    print(f"Failed items: {failed_profiles}")
```
